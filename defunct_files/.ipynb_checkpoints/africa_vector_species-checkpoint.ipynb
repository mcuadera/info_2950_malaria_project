### A dataset of dominant malaria vector and location

```python
import pandas as pd
import matplotlib.pyplot as plt
bionomics_africa = pd.read_csv('bionomics_africa.csv', low_memory=False)
```

```python
#removing all but a few columns of the csv file
bionomics_africa = bionomics_africa[['country', 'lat', 'long', 'insecticide_control', 'season_given', 'species']]
print(bionomics_africa.columns)
```

```python
#All of the different species of viral vector mosquitoes 
uniques = bionomics_africa.species.unique()
print(uniques)
```

```python
#plotting the data
plt.scatter(bionomics_africa.country, bionomics_africa.species)
plt.show()
```
