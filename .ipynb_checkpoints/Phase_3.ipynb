# Phase III: Preregistration of Analyses

Overview: our group performed many analyses for our Phase II submission and hope to include those in our final project. The below analyses are more two that we did not perform for Phase II and are more complicated and in-depth. For these, we will include two variables in our dataset that were not present in our earlier submission: temperature and GDP by country per year. We may change these variables if it's not possible to find the right data and will communicate that to the TAs. Our goal is to determine whether countries are more likely to have a higher incidence of malaria given certain characteristics, such as average temprature, region, GDP, and others.

### Analysis 1: Multi-Variable Regression of Temperature, GDP and Malaria Incidence 
We will create a multivariable linear regression that predicts the incidence of malaria given temperature and GDP by country for 2017 (the most recent year in our dataset). Temperature and GDP by year and country were not variables in the dataset we used for Phase II, but we will perform web scraping to add those variables to the final dataset. We will observe the resulting coefficient of the model and draw conclusions based on that. For instance, if the coefficient is large and positive, we can infer that temperature and GDP together correlate with a higher incidence of malaria in most countries. From the evidence we gather, we will discuss the potential impact of global warming on malaria incidence worldwide.

### Analysis 2: K-Means
We will perform a k-means clustering analysis to discover ways in which the countries in the dataset can be categorized. We will try different clustering with different values of k. A significant result would be that countries with high, medium and low malaria incidence rates share common categoristics, such as region or temperature change. This analysis will provide further insight as to what country characteristics may put countries at a higher risk for the spread of malaria.
